```python
class Solution:
    def waysToSplit(self, nums) -> int:
        l = 0
        r = len(nums) - 1
        ct = 0
        while l <= r:
            mid = (l + r) // 2
            ls = sum(nums[:mid])
            rs = sum(nums[mid:])
            if ls <= nums[mid] <= rs:
                ct += 1
            if ls < rs:
                l += 1
            else:
                r -= 1
        return ct

s = Solution()
print(s.waysToSplit([1,1,1]))
```
